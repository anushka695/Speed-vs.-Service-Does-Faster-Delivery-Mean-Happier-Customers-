```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Fast Delivery Agent Reviews.csv")

# 1. Customer Service Rating vs. Rating
service_rating_analysis = df.groupby('Customer Service Rating')['Rating'].agg(['mean', 'count']).reset_index()
service_rating_analysis.columns = ['Customer Service Rating', 'Avg. Customer Rating', 'Count']

# 2. Order Accuracy vs. Rating
order_accuracy_analysis = df.groupby('Order Accuracy')['Rating'].agg(['mean', 'count']).reset_index()
order_accuracy_analysis.columns = ['Order Accuracy', 'Avg. Customer Rating', 'Count']

# 3. Product Availability vs. Rating
product_availability_analysis = df.groupby('Product Availability')['Rating'].agg(['mean', 'count']).reset_index()
product_availability_analysis.columns = ['Product Availability', 'Avg. Customer Rating', 'Count']

# 4. Price Range vs. Rating
price_range_analysis = df.groupby('Price Range')['Rating'].agg(['mean', 'count']).reset_index()
price_range_analysis.columns = ['Price Range', 'Avg. Customer Rating', 'Count']

# 5. Discount Applied vs. Rating
discount_analysis = df.groupby('Discount Applied')['Rating'].agg(['mean', 'count']).reset_index()
discount_analysis.columns = ['Discount Applied', 'Avg. Customer Rating', 'Count']

# Print out the analysis results
print("Customer Service Rating vs. Rating Analysis:")
print(service_rating_analysis)
print("\nOrder Accuracy vs. Rating Analysis:")
print(order_accuracy_analysis)
print("\nProduct Availability vs. Rating Analysis:")
print(product_availability_analysis)
print("\nPrice Range vs. Rating Analysis:")
print(price_range_analysis)
print("\nDiscount Applied vs. Rating Analysis:")
print(discount_analysis)
```
