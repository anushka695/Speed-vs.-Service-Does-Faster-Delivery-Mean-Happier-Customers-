```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("Fast Delivery Agent Reviews.csv")

# Display first few rows
print(df.head())

# Convert ratings and delivery time to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Delivery Time'] = pd.to_numeric(df['Delivery Time'], errors='coerce')

# Drop missing values
df.dropna(subset=['Rating', 'Delivery Time'], inplace=True)

# Summary statistics
print(df[['Rating', 'Delivery Time']].describe())

# Correlation analysis
correlation, p_value = pearsonr(df['Delivery Time'], df['Rating'])
print(f"Correlation: {correlation}, P-value: {p_value}")

# Scatter plot: Delivery Time vs. Ratings
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Delivery Time'], y=df['Rating'])
plt.title("Delivery Time vs Customer Ratings")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Customer Rating")
plt.show()

# Regression analysis
X = sm.add_constant(df['Delivery Time'])  # Add constant for intercept
y = df['Rating']
model = sm.OLS(y, X).fit()
print(model.summary())

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("Fast Delivery Agent Reviews.csv")

# Display first few rows
print(df.head())

# Convert ratings and delivery time to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Delivery Time (min)'] = pd.to_numeric(df['Delivery Time (min)'], errors='coerce')

# Drop missing values
df.dropna(subset=['Rating', 'Delivery Time (min)'], inplace=True)

# Summary statistics
print(df[['Rating', 'Delivery Time (min)']].describe())

# Correlation analysis
correlation, p_value = pearsonr(df['Delivery Time (min)'], df['Rating'])
print(f"Correlation: {correlation}, P-value: {p_value}")

# Scatter plot: Delivery Time vs. Ratings
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Delivery Time (min)'], y=df['Rating'])
plt.title("Delivery Time vs Customer Ratings")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Customer Rating")
plt.show()

# Regression analysis
X = sm.add_constant(df['Delivery Time (min)'])  # Add constant for intercept
y = df['Rating']
model = sm.OLS(y, X).fit()
print(model.summary())
```
